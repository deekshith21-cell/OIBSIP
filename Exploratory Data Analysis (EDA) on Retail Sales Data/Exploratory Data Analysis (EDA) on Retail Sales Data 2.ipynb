```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")

sns.set(style="whitegrid")
```

```python
# Load your dataset
df = pd.read_csv(r"C:\Users\Dell\Downloads\menu.csv")

# Show the first few rows
df.head()
```

```python
print("Dataset shape:", df.shape)
print("\nColumns:\n", df.columns)
print("\nData types:\n", df.dtypes)
print("\nMissing values:\n", df.isnull().sum())
print("\nDuplicate records:", df.duplicated().sum())
```

```python
df.describe()
```

```python
top_cal = df[['Item', 'Calories']].sort_values(by='Calories', ascending=False).head(10)

plt.figure(figsize=(8, 5))
sns.barplot(data=top_cal, x='Calories', y='Item', palette='Reds_r')
plt.title("Top 10 Highest-Calorie Menu Items")
plt.xlabel("Calories")
plt.tight_layout()
plt.show()
```

```python
top_sodium = df[['Item', 'Sodium']].sort_values(by='Sodium', ascending=False).head(10)

plt.figure(figsize=(8, 5))
sns.barplot(data=top_sodium, x='Sodium', y='Item', palette='Blues_r')
plt.title("Top 10 High-Sodium Menu Items")
plt.xlabel("Sodium (mg)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='YlGnBu', fmt=".1f")
plt.title("Correlation Between Nutritional Components")
plt.show()
```

```python
df.groupby('Category')['Calories'].mean().sort_values(ascending=False).plot(kind='bar', figsize=(8, 5), color='orange')
plt.title("Average Calories by Food Category")
plt.ylabel("Calories")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
